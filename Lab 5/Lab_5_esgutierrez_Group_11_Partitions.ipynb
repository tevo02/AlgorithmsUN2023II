# Write a code that produces the the ** Bell numbers** count the possible partitions of a set.

**Hint**:
. Output the n-th Bell Number - [stackexchange](https://codegolf.stackexchange.com/questions/132379/output-the-n-th-bell-number)

```python
def bell(n, k=0):
    if n < 1:
        return 1
    return k * bell(n-1, k) + bell(n-1, k+1)

# Test cases
for i in range(16):
    print(f"{i} -> {bell(i)}")
```

###Write a code that that prduces  list of all partititions of the set of $[n]=\{ 1,2,3,4\cdots , n-1, n\}$ with $0 \le n \le 65$ and also list all the partitions for which the sum of elements of the partition is the same constant for all.

```python
lista_sumas_iguales=[]

def print_partition(ans):
    """
    Function to print a partition
    """
    for i in ans:
        print("{", end=" ")
        for element in i:
            print(element, end=" ")
        print("}", end=" ")
    print()


def partition_set(set, index, ans):
    """
    Function to generate all partitions
    """
    if index == len(set):
        # If we have considered all elements in the set, print the partition
        global lista_sumas_iguales
        print_partition(ans)
        lista_s=[]
        lista_sumas=[]
        for lista in ans:
          lista_s.append(sum(lista))
          lista_sumas.append(lista)
        if len(lista_s)>1:
          if all(x == lista_s[0] for x in lista_s):
            for sublista in lista_sumas:
              lista_sumas_iguales.append(sublista.copy())
            print("En la particion anterior La suma de los elemetos es la misma.")

        return

    # For each subset in the partition, add the current element to it and recall
    for i in range(len(ans)):
        ans[i].append(set[index])
        partition_set(set, index + 1, ans)
        ans[i].pop()

    # Add the current element as a singleton subset and recall
    ans.append([set[index]])
    partition_set(set, index + 1, ans)
    ans.pop()


def all_partitions(set):
    """
    Function to generate all partitions for a given set
    """
    ans = []
    partition_set(set, 0, ans)


# Main function
if __name__ == "__main__":
    # The size of the set
    n= int(input("Escribe un numero entre 1 y 65: "))

    # Initialize the set as {1, 2, ..., n}
    set = list(range(1, n + 1))

    print("All partitions of the set will be:")
    # Generate all partitions of the set
    all_partitions(set)

    print(f"las particiones con las cuales la suma de los elementos es la misma es: {lista_sumas_iguales}")
```

# Write a code that produces the partitions with all the parts having the same sum value, identify the formula for $n < 16$.

```python
def constant_partition(n):
    def partition_util(n, m, result):
        if n == 0:
            return [[]]
        if n < 0 or m == 0:
            return []
        res = []
        for i in range(1, m+1):
            for p in partition_util(n-i, i, result):
                res.append([i] + p)
        return res

    result = []
    partitions = partition_util(n, n, result)
    constant_sum_partitions = [p for p in partitions if len(set(p)) == 1]
    return constant_sum_partitions

# Test cases for constant partitions with the same sum
for i in range(16):
    partitions = constant_partition(i)
    if partitions:
        print(f"Constant sum partitions for {i}: {partitions}")
```
