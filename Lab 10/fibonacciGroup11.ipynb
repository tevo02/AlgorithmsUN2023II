```python
import time
import matplotlib.pyplot as plt

def recursive_fibonacci(n):
    if n <= 1:
        return n
    return recursive_fibonacci(n-1) + recursive_fibonacci(n-2)

def iterative_fibonacci_linear(n):
    if n <= 1:
        return n
    fib = [0, 1]
    for i in range(2, n+1):
        fib.append(fib[-1] + fib[-2])
    return fib[n]

def iterative_fibonacci_logarithmic(n):
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(n-1):
        a, b = b, a + b
    return b


recursive_times = []

for i in range(51):
    start_time = time.time()
  # recursive_fibonacci(51)
    end_time = time.time()
    execution_time = end_time - start_time
    recursive_times.append(execution_time)

iterative_linear_times = []
for i in range(1001):
    start_time = time.time()
    iterative_fibonacci_linear(i)
    end_time = time.time()
    execution_time = end_time - start_time
    iterative_linear_times.append(execution_time)

iterative_logarithmic_times = []
for i in range(1001):
    start_time = time.time()
    iterative_fibonacci_logarithmic(i)
    end_time = time.time()
    execution_time = end_time - start_time
    iterative_logarithmic_times.append(execution_time)

x_recursive = list(range(51))
x_iterative = list(range(1001))
y_n = x_iterative

plt.plot(x_recursive, recursive_times, label='Recursive Fibonacci')
plt.plot(x_iterative, iterative_linear_times, label='Iterative Fibonacci (Linear)')
plt.plot(x_iterative, iterative_logarithmic_times, label='Iterative Fibonacci (Logarithmic)')
plt.plot(x_iterative, y_n, label='y = n')

plt.xlabel('n')
plt.ylabel('Execution Time (seconds)')
plt.title('Fibonacci Execution Time Comparison')
plt.legend()
plt.grid(True)
plt.show()
```
